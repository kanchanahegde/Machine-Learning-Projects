```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
import pickle
```

```python
with open('Dataset/train_X.p', 'rb') as f:
    train_X = pickle.load(f)
train_X.head()
```

```python
with open('Dataset/train_y.p', 'rb') as f:
    train_y= pickle.load(f)
train_y.head()
```

```python
with open('Dataset/test_X.p', 'rb') as f:
    test_X = pickle.load(f)
test_X.head()
```

```python
with open('Dataset/test_y.p', 'rb') as f:
    test_y = pickle.load(f)
test_y.head()
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_X.values,train_y.values)
y_pred= knn.predict(test_X.values)
print("acc",accuracy_score(test_y.values,y_pred))
```

```python
from sklearn.decomposition import PCA
n_components =60
pca = PCA(n_components=n_components, svd_solver='randomized',whiten=True).fit(train_X)
X_train_pca = pca.transform(train_X)
train_new=np.array(X_train_pca)
X_test_pca = pca.transform(test_X)
test_new=np.array(X_test_pca)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train_new,train_y.values)
y_pred= knn.predict(test_new)
print("acc",accuracy_score(test_y.values,y_pred))
```

```python
X_train_pca = pca.transform(train_X)
train_new=np.array(X_train_pca)
X_test_pca = pca.transform(test_X)
test_new=np.array(X_test_pca)
```

```python
kvals=range(1,15,2)
accuracy=[]
n_components = [20,40,60,80]
for k in kvals:
    for nc in n_components:
       train_new = PCA(n_components=nc, svd_solver='randomized',whiten=True).fit_transform(train_X)
       knn = KNeighborsClassifier(n_neighbors=k)
       score = cross_val_score(knn,train_new,train_y,cv=10,scoring='accuracy')
       accuracy.append(score.mean()*100)
       print(k,nc,score.mean()*100)
    
```
